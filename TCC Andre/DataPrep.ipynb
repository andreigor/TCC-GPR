```python
cepx = list(ceps.Cep.values)
lats = list(ceps.latitudes.values)
longs = list(ceps.longitudes.values)

cep_dict = {cep:(lat,longs) for cep, lat, longs in list(zip(cepx,lats,longs))}

```

```python
def put_lat(cep):
    lat = cep_dict[cep][0]
    return lat
def put_long(cep):
    long = cep_dict[cep][1]
    return long

```

```python
df_cg['lattt'] = df_cg['cep'].apply(put_lat)
df_cg['longgg'] = df_cg['cep'].apply(put_long)

df_cg.head()
```
